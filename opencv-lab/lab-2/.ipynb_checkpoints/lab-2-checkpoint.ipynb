```python
import cv2 
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("butterfly.png")
```

```python
print(img.dtype)
```

```python
print(img.shape)
```

```python
print(img[0,0])
```

```python
plt.imshow(img)
```

```python
from cv2 import cvtColor,COLOR_BGR2RGB,COLOR_RGB2HSV,COLOR_RGB2LAB
```

```python
img_rgb=cvtColor(img,COLOR_BGR2RGB)
```

```python
print(img_rgb[0,0])
```

```python
plt.imshow(img_rgb)
```

```python
r,g,b=cv2.split(img_rgb)
```

```python
plt.figure(figsize=(10,10)) 
```

```python
plt.subplot(221)
plt.title("Original")
plt.imshow(img_rgb)
plt.subplot(222)
plt.title("Blue Channel")
plt.imshow(b)
plt.subplot(223)
plt.title("green channel")
plt.imshow(g)
plt.subplot(224)
plt.title("red channel")
plt.imshow(r)
```
