```python
import matplotlib.pyplot as plt
import os
import numpy as np
import cv2
```

# 1. Reading,Displaying and Writing an image

```python
img = cv2.imread('random.png')

# Displaying the image using Matplotlib
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Normal Image')
plt.axis('off')  
plt.show()

# Writing the image
if not os.path.exists('output'):
    os.makedirs('output')
cv2.imwrite(os.path.join('output', 'image.png'), img)
```

# 2. Changing the format from png to jpg

```python
cv2.imwrite(os.path.join('output', 'image.jpg'), img)
```

# 3. Resizing the Image

```python
img_resized = cv2.resize(img, (1050, 1000))
plt.imshow(cv2.cvtColor(img_resized, cv2.COLOR_BGR2RGB))
plt.title('Resized Image')
plt.axis('off')  
plt.show()
```

# 4. Converting the image to grayscale

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
plt.imshow(cv2.cvtColor(gray_image, cv2.COLOR_BGR2RGB))
plt.title('Grayscale Image')
plt.axis('off')  
plt.show()
```

# 6.1 - Playing a video

```python
cap = cv2.VideoCapture('video.mp4')

if not cap.isOpened():
    print("Error opening video file")

while cap.isOpened():
    ret, frame = cap.read()
    if ret:
        cv2.imshow('Frame', frame)
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()
```

# 6.2- Extracting and saving all the frames of the video

```python
cam = cv2.VideoCapture('video.mp4')

try:
    if not os.path.exists('data'):
        os.makedirs('data')
except OSError:
    print('Error: Creating directory of data')

currentframe = 0

while True:
    ret, frame = cam.read()

    if ret:
        name = os.path.join('data', 'frame' + str(currentframe) + '.jpg')
        print('Creating...' + name)
        cv2.imwrite(name, frame)
        currentframe += 1
    else:
        break

cam.release()
cv2.destroyAllWindows()
```
